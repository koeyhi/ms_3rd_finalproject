# 특성 공학(Feature Engineering)

## 날짜 형식으로 변환

```python
train_tr["구매일시"] = pd.to_datetime(train_tr["구매일시"])
test_tr["구매일시"] = pd.to_datetime(test_tr["구매일시"])
```

## 새로 만든 feature와 병합할 고객ID로만 이루어진 데이터프레임 생성

```python
train_ft = train_target[["ID"]]
test_ft = submit[["ID"]]

train_ft.shape, test_ft.shape
```

## 구매일시를 이용한 특성생성

```python
agg_list = [
    ("내점일수", "nunique"),
    ("주말방문비율", lambda x: np.mean(x.dt.weekday>4)),
    ("구매주기", lambda x: int((x.max() - x.min()).days / x.dt.date.nunique())),
    ("거래개월수", lambda x: x.dt.date.astype(str).str[:-3].nunique()),
    # 일별 월별 연도별 평균 구매건수
    ("일별평균구매건수", lambda x: x.count() / x.dt.date.nunique()),
    ("월별평균구매건수", lambda x: x.count() / x.dt.month.nunique()),  
    ("연도별평균구매건수", lambda x: x.count() / x.dt.year.nunique()),
    ("주구매요일", lambda x: x.dt.weekday.mode()[0]),
    # 요일별
    ("월요일_구매비율", lambda x: np.mean(x.dt.weekday == 0)),
    ("화요일_구매비율", lambda x: np.mean(x.dt.weekday == 1)),
    ("수요일_구매비율", lambda x: np.mean(x.dt.weekday == 2)),
    ("목요일_구매비율", lambda x: np.mean(x.dt.weekday == 3)),
    ("금요일_구매비율", lambda x: np.mean(x.dt.weekday == 4)),
    ("토요일_구매비율", lambda x: np.mean(x.dt.weekday == 5)),
    ("일요일_구매비율", lambda x: np.mean(x.dt.weekday == 6)),
    #월별
    ("1월_구매비율", lambda x: np.mean(x.dt.month == 1)),
    ("2월_구매비율", lambda x: np.mean(x.dt.month == 2)),
    ("3월_구매비율", lambda x: np.mean(x.dt.month == 3)),
    ("4월_구매비율", lambda x: np.mean(x.dt.month == 4)),
    ("5월_구매비율", lambda x: np.mean(x.dt.month == 5)),
    ("6월_구매비율", lambda x: np.mean(x.dt.month == 6)),
    ("7월_구매비율", lambda x: np.mean(x.dt.month == 7)),
    ("8월_구매비율", lambda x: np.mean(x.dt.month == 8)),
    ("9월_구매비율", lambda x: np.mean(x.dt.month == 9)),
    ("10월_구매비율", lambda x: np.mean(x.dt.month == 10)),
    ("11월_구매비율", lambda x: np.mean(x.dt.month == 11)),
    ("12월_구매비율", lambda x: np.mean(x.dt.month == 12)),
    #계절별
    ("봄_구매비율")
]

tmp = train_tr.groupby("ID")["구매일시"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how="left", on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby("ID")["구매일시"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how="left", on="ID")
test_ft.head()
```

```python
# 주말 비율 추가
agg_list = [
    ("평일_구매비율", lambda x: np.mean(x.dt.weekday <= 4)),
    ("주말_구매비율", lambda x: np.mean(x.dt.weekday >= 5)),
]

tmp = train_tr.groupby("ID")["구매일시"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how="left", on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby("ID")["구매일시"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how="left", on="ID")
test_ft.head()
```

```python
agg_list = [
    ("봄_구매비율", lambda x: np.mean((x.dt.month >= 3) & (x.dt.month <= 5))),
    ("여름_구매비율", lambda x: np.mean((x.dt.month >= 6) & (x.dt.month <= 8))),
    ("가을_구매비율", lambda x: np.mean((x.dt.month >= 9) & (x.dt.month <= 11))),
    ("겨울_구매비율", lambda x: np.mean((x.dt.month == 12) | (x.dt.month <= 2))),
]

tmp = train_tr.groupby("ID")["구매일시"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how="left", on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby("ID")["구매일시"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how="left", on="ID")
test_ft.head()
```

```python
train_ft.to_csv(f"{DATA_PATH}train_common.csv",index=False)
test_ft.to_csv(f"{DATA_PATH}test_common.csv",index=False)
```

## 구매가격을 이용한 특성생성

```python
train_ft = pd.read_csv(f"{DATA_PATH}train_common.csv")
test_ft = pd.read_csv(f"{DATA_PATH}test_common.csv")
```

```python
train_tr["구매가격"].quantile(0.75)
```

- 구매가격 0 이상중 4분위수 계산

```python
high = train_tr.loc[train_tr["구매가격"] > 0, "구매가격"].quantile(0.75)
low = train_tr.loc[train_tr["구매가격"] > 0, "구매가격"].quantile(0.25)

high, low
```

```python
low_refund = train_tr.loc[train_tr["구매가격"] < 0, "구매가격"].quantile(0.75)
high_refund = train_tr.loc[train_tr["구매가격"] < 0, "구매가격"].quantile(0.25)

high_refund, low_refund
```

```python
agg_list = [
    ("총구매액_환불포함", "sum"),
    ("총구매액", lambda x: x[x > 0].sum()),
    ("총구매건수", lambda x: x[x > 0].count()),
    ("평균구매금액", lambda x: x[x > 0].mean()),
    ("고액구매비율", lambda x: np.mean(x[x > 0] >= high)),
    ("저액구매비율", lambda x: np.mean(x[x > 0] <= low)),
    ("최대구매금액", lambda x: x[x > 0].max()),
    ("최소구매금액", lambda x: x[x > 0].min()),
    ("구매금액표준편차", lambda x: x[x > 0].std()),
    ("구매금액_skew", lambda x: x[x > 0].skew()),
    ("구매금액_kurtosis", lambda x: x[x > 0].kurtosis()),
    ("총환불금액", lambda x: x[x < 0].sum()),
    ("환불건수", lambda x: (x < 0).sum()),
    ("고액환불비율", lambda x: np.mean(x[x < 0] <= high_refund)),
    ("저액환불비율", lambda x: np.mean(x[x < 0] >= low_refund)),
    ("환불금액표준편차", lambda x: x[x < 0].std()),
    ("환불금액_skew", lambda x: x[x < 0].skew()),
    ("환불금액_kurtosis", lambda x: x[x < 0].kurtosis()),
    ("최소환불금액", lambda x: x[x < 0].max()),
    ("최대환불금액", lambda x: x[x < 0].min()),
]

tmp = train_tr.groupby("ID")["구매가격"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how="left")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["구매가격"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left')
test_ft.head()
```

```python
train_ft.to_csv(f"{DATA_PATH}train_common.csv",index=False)
test_ft.to_csv(f"{DATA_PATH}test_common.csv",index=False)
```

## pivot_table을 이용한 특성 생성

- 지점코드별

```python
purchase_pivot = (
    test_tr[test_tr["구매가격"] > 0]
    .groupby(["ID", "지점코드"])
    .size()
    .unstack(fill_value=0)
    .add_prefix("구매지점_")
)

refund_pivot = (
    test_tr[test_tr["구매가격"] < 0]
    .groupby(["ID", "지점코드"])
    .size()
    .unstack(fill_value=0)
    .add_prefix("환불지점_")
)

```

- 중분류별

```python
purchase_pivot_2 = (
    test_tr[test_tr["구매가격"] > 0]
    .groupby(["ID", "중분류"])
    .size()
    .unstack(fill_value=0)
    .add_prefix("중분류_")
)

refund_pivot_2 = (
    test_tr[test_tr["구매가격"] < 0]
    .groupby(["ID", "중분류"])
    .size()
    .unstack(fill_value=0)
    .add_prefix("중분류_")
)

purchase_pivot = purchase_pivot.merge(purchase_pivot_2, how="left", on="ID")
refund_pivot = refund_pivot.merge(refund_pivot_2, how="left", on="ID")

purchase_pivot
```

```python
purchase_pivot_3 = (
    test_tr[test_tr["구매가격"] > 0]
    .groupby(["ID", "대분류"])
    .size()
    .unstack(fill_value=0)
    .add_prefix("대분류_")
)

refund_pivot_3 = (
    test_tr[test_tr["구매가격"] < 0]
    .groupby(["ID", "대분류"])
    .size()
    .unstack(fill_value=0)
    .add_prefix("대분류_")
)
```

```python
purchase_pivot = purchase_pivot.merge(purchase_pivot_3, how="left", on="ID")
refund_pivot = refund_pivot.merge(refund_pivot_3, how="left", on="ID")
```

### 최종 pivot table merge

```python
# train_ft = train_ft.merge(purchase_pivot, on="ID", how="left")
# train_ft = train_ft.merge(refund_pivot, on="ID", how="left")
test_ft = test_ft.merge(purchase_pivot, how="left", on="ID")
test_ft = test_ft.merge(refund_pivot, how="left", on="ID")
```

```python
train_ft.to_csv(f"{DATA_PATH}train_common.csv",index=False)
test_ft.to_csv(f"{DATA_PATH}test_common.csv",index=False)
```
