## 날짜 형식으로 변환

```python
train_tr["구매일시"] = pd.to_datetime(train_tr["구매일시"])
test_tr["구매일시"] = pd.to_datetime(test_tr["구매일시"])
```

## 구매일시를 이용한 특성생성

```python
agg_list = [
        # 컬럼명, 집계 방식
        ('내점일수','nunique'),
        ('구매주기', lambda x: int( (x.max() - x.min()).days / x.dt.date.nunique()) ),
        ('주말방문비율', lambda x: np.mean(x.dt.weekday>4)),
        ('봄_구매비율', lambda x: np.mean(x.dt.month.isin([3,4,5]))),
        ('여름_구매비율', lambda x: np.mean(x.dt.month.isin([6,7,8]))),
        ('가을_구매비율', lambda x: np.mean(x.dt.month.isin([9,10,11]))),
        ('겨울_구매비율', lambda x: np.mean(x.dt.month.isin([1,2,12]))),
        ('주구매요일', lambda x: x.dt.weekday.mode()[0]),
        ('일별평균구매건수', lambda x:  x.count() / x.dt.date.nunique() ),
        ('거래개월수', lambda x: x.dt.date.astype(str).str[:-3].nunique() ),

        # 구매 월
        ('1월 구매 비율', lambda x : np.mean(x.dt.month == 1)),
        ('2월 구매 비율', lambda x : np.mean(x.dt.month == 2)),
        ('3월 구매 비율', lambda x : np.mean(x.dt.month == 3)),
        ('4월 구매 비율', lambda x : np.mean(x.dt.month == 4)),
        ('5월 구매 비율', lambda x : np.mean(x.dt.month == 5)),
        ('6월 구매 비율', lambda x : np.mean(x.dt.month == 6)),
        ('7월 구매 비율', lambda x : np.mean(x.dt.month == 7)),
        ('8월 구매 비율', lambda x : np.mean(x.dt.month == 8)),
        ('9월 구매 비율', lambda x : np.mean(x.dt.month == 9)),
        ('10월 구매 비율', lambda x : np.mean(x.dt.month == 10)),
        ('11월 구매 비율', lambda x : np.mean(x.dt.month == 11)),
        ('12월 구매 비율', lambda x : np.mean(x.dt.month == 12)),

        # 구매 시간
        ('오전 구매 비율', lambda x : np.mean(x.dt.hour < 12)), ###
        ('오후 구매 비율', lambda x : np.mean(x.dt.hour >= 12)),###
        ('주구매시간대', lambda x: x.dt.hour.mode()[0]),  # 가장 자주 구매한 시간대(시간)
        ('평균구매_시간', lambda x: x.dt.hour.mean()),    # 평균 구매 시간
        ('야간구매비율', lambda x: np.mean((x.dt.hour >= 18) | (x.dt.hour <= 6))),  # 야간 구매 비율 18시 ~ 05시


        # 구매 요일 특성
        ('평균구매_요일', lambda x: x.dt.weekday.mean()),        # 평균 구매 요일
        ('주중 구매 비율', lambda x : np.mean(x.dt.weekday < 5)),# 주중 구매 비율(월~금 0,1,2,3,4)
        ('주말 구매 비율', lambda x : np.mean(x.dt.weekday >= 5)),



        #
        #('구매간격', lambda x: (x.max() - x.min()).days),  # 첫 구매와 마지막 구매 사이의 기간
        #('구매일간격', lambda x: np.mean(np.diff(x.dt.date.astype(int)))),  # 평균 구매 사이의 일수
        #('할인상품_구매비율', lambda x: np.mean(x.str.contains('할인', regex=True))),  # 할인 상품 비율
        #('최근구매일', lambda x: (pd.to_datetime('today') - x.max()).days),  # 마지막 구매일로부터의 경과 일수

]

tmp = train_tr.groupby('ID')["구매일시"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["구매일시"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
test_ft.head()
```

## 브랜드코드를 이용한 특성생성

```python
agg_list = [
             ('브랜드코드_nunique', 'nunique'),
             ('브랜드구매비율', lambda x: x.value_counts(normalize=True).get(x.mode()[0], 0)),  # 주 구매 브랜드 비율  ###
             ('주구매브랜드', lambda x: x.mode()[0]),  # 가장 자주 구매한 브랜드  ###
             ('브랜드별구매횟수', 'count'),  # 총 구매 횟수  ###
             ('브랜드별평균구매액', 'mean'),  # 평균 구매 금액 ###
             ('브랜드별최대구매액', 'max'),  # 최대 구매 금액 ###
             ('브랜드다양성', 'nunique'),  # 구매한 브랜드의 수 ###



             ]

tmp = train_tr.groupby('ID')["브랜드코드"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
train_ft.head()
```

## 중분류를 이용한 특성생성

```python
#미혼 기혼 구입항목에서 차이가 있을거라 생각하는 품목
agg_list = [
    ('중분류_nunique', 'nunique'),
    ('주구매_중분류', lambda x: x.mode()[0]),
    ('주구매_중분류_아동구매비율', lambda x: (x.str.contains("아동|유아|신생아|완구|팬시", regex=True)).mean()), #포함 문자열만 추출
    ('주구매_중분류_취미비율', lambda x: (x.str.contains('아웃도어|골프|스포츠|취미|수예', regex=True)).mean()),
    ('주구매_중분류_캐릭터비율', lambda x: (x.str.contains('캐릭터', regex=True)).mean()),
    ('주구매_중분류_가전제품비율', lambda x: (x.str.contains('TV|세탁기|냉장고', regex=True)).mean()),
    ('주구매_중분류_가구비율', lambda x: (x.str.contains('침대|침구|가구|쇼파|식탁', regex=True)).mean()),
    ('주구매_중분류_고가비율', lambda x: (x.str.contains('보석|모피,', regex=True)).mean()),
]

tmp = train_tr.groupby('ID')["중분류"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left', on='ID')  # ID를 기준으로 병합
train_ft.head()
```
