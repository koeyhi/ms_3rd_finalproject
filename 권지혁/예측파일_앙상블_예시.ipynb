# 팀원들의 예측 파일을 평균 내는 등과 같이 앙상블을 할 경우
- 팀원들에게 csv 파일 만들때 1에 대한 확률로 예측할 것을 요구해야 함

```python
DATA_PATH = "/content/drive/MyDrive/파이널프로젝트/data/"
```

```python
import pandas as pd
import numpy as np

boss = pd.read_csv(f"{DATA_PATH}karns_submit.csv") # 이관수 팀장의 예측 파일이라고 가정
gildong = pd.read_csv(f"{DATA_PATH}gildong_submit.csv") # 홍길동 팀원의 예측 파일이라고 가정
minsu = pd.read_csv(f"{DATA_PATH}minsu_submit.csv") # 김민수 팀원의 예측 파일이라고 가정
```

```python
pred = np.mean(
        [
            boss["target"].to_numpy(),
            gildong["target"].to_numpy(),
            minsu["target"].to_numpy()
        ],axis=0)
pred.shape
```

```python
pred = (pred >= 0.5).astype(int)
pred
```

```python
submit = boss[["ID"]]
submit["target"] = pred
submit
```

```python
submit.to_csv(f"{DATA_PATH}ensemble_boss_gildong_minsu.csv",index=False)
```
